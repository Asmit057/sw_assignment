```python
# Q3.. What errors are defined in the ArithmeticError class? Explain any two with an example.
```

```python
# The ArithmeticError Exception is the base class for all errors associated with arithmetic operation.
# errors are:
#  ZeroDivisionError
# Handling OverflowError
```

```python
# ZeroDivisionError
# it occur when we divide a number by 0.
try:
    1/0
except ArithmeticError as e:
    print(f"{e}, {e.__class__}")
```

```python
# Handling OverflowError
# An OverflowError exception is raised when an arithmetic operation exceeds the limits to be represented.
j = 5.0

try:
    for i in range(1, 1000):
        j = j**i
        print(j)
except OverflowError as e:
    print("Overflow error happened")
    print(f"{e}, {e.__class__}")
```

```python
# Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
```

```python
# The LookupError exception in Python forms the base class for all exceptions that are raised when
#  an index or a key is not found for a sequence or dictionary respectively.
# keyerror
# A KeyError exception is raised when you are trying to access an invalid key in a python dictionary.
try :
    d = {"key" :"aman" , 1 : [2,3,4,5]}
    print(d["key2"])
except KeyError as e : 
    print(e)
```

```python
# IndexError
# It is used when index of object is out of range
try:
    list1 = [1,2,3,4,5]
    list1[12]
except IndexError as e:
    print("index error is ",e)
```
